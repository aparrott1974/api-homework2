```python
import pandas as pd
import numpy 
```

```python
coords=pd.read_csv("resources/worldcitiespop.csv")
coords.head()
```

```python
print(len(coords))
```

```python
#drop duplicate values except the last one
ucoords=coords.drop_duplicates(['Latitude','Longitude']) #, keep='last')
ucoords=ucoords.drop(columns=['AccentCity', 'Region'])
ucoords.head()
```

```python
print(len(ucoords))
```

```python
ucoords.dtypes
```

```python
# get only the ucoords with >499 pop
#  df = df[df.line_race != 0]
ucoords = ucoords[ucoords.Population > .5]
ucoords.head()
```
